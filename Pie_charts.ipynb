```python
import Dominion
import matplotlib.pyplot as plt
import random
import pandas
```

```python
n=3 #number of players
g=500 #number of games
Player_list = ['Adam','Bonnie','Chelsea'][:n]#player names without prefixes
autoplayers=['*Adam','*Bonnie','*Chelsea']#player names with prefixes
Wins = {}
```

```python
#Play g games, and note the winners for each. Compare % of wins by player order, or by different computer players
for p in Player_list:
    Wins[p]=0
for i in range(g):
    random.shuffle(autoplayers)
    winners=Dominion.playgame(autoplayers,True)[0]
    for w in winners:
        Wins[w]+=1
```

```python
#Pie chart showing win share by player. Note that one game can have more than one winner.
#So the chart shows share of wins by player, not the % of games won by that player.
fig1,ax1=plt.subplots()
sizes=[Wins[p] for p in Player_list]
ax1.pie(sizes,labels=Player_list,autopct='%1.0f%%')
plt.title('Win share, first 3 auto-players, ' + str(int(n)) + ' player Dominion')
plt.show()
```
