```python
import Dominion
import matplotlib.pyplot as plt
```

```python
n=2 #number of players
g=100 #number of games
Player_list = ['TP1','CP2','*Player 3','*Player 4'][:n]#player names without prefixes
autoplayers=['^TP1','*CP2']#player names with prefixes
Wins = {}
```

```python
#Play g games, and note the winners for each. Compare % of wins by player order, or by different computer players
for p in Player_list:
    Wins[p]=0
for i in range(g):
    winners=Dominion.playgame(autoplayers,True)[0]
    for w in winners:
        Wins[w]+=1
```

```python
#Pie chart showing win share by player. Note that one game can have more than one winner.
#So the chart shows share of wins by player, not the % of games won by that player.
fig1,ax1=plt.subplots()
sizes=[Wins[p] for p in Player_list]
ax1.pie(sizes,labels=Player_list,autopct='%1.0f%%')
plt.title('Win percentages, computer vs table player, ' + str(int(n)) + ' player Dominion')
plt.show()
```
