```python
import Dominion
import matplotlib.pyplot as plt
```

```python
n=2 #number of players
g=1000 #number of games
Player_list = ['Player 1','Player 2','Player 3','Player 4']
sublist = Player_list[:n]
autoplayers = ['*'+p for p in sublist]
Wins = {}
```

```python
for p in sublist:
    Wins[p]=0
for i in range(g):
    winners=Dominion.playgame(autoplayers,True)
    for w in winners:
        Wins[w]+=1
```

```python
fig1,ax1=plt.subplots()
sizes=[Wins[p] for p in sublist]
ax1.pie(sizes,labels=sublist,autopct='%1.0f%%')
plt.title('Win percentages by player order, ' + str(int(n)) + ' player Dominion')
plt.show()
```
